```python
import numpy as np
import pandas as pd
from tensorflow.keras import layers, models, metrics, losses
from sklearn.model_selection import train_test_split
import tensorflow_addons as tfa
import matplotlib.pyplot as plt
```

```python
features = pd.read_csv('features.csv')
target = pd.read_csv('target.csv')
```

```python
features = features.iloc[:, features.columns != 'Unnamed: 0']
```

```python
features
```

```python
target = target.iloc[:, target.columns != 'Unnamed: 0']
```

```python
target
```

```python
# split data into train/test
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.1, random_state=0)
```

```python
X_train.shape, y_train.shape
```

```python
X_train.values
```

```python
X_train.values.shape
```

```python
xTrainLast = np.expand_dims(X_train.values, axis=2)
xTrainLast.shape
```

```python
xTestLast = np.expand_dims(X_test.values, axis=2)
xTestLast.shape
```

```python
yTrainFlat = y_train.values.flatten()
yTestFlat = y_test.values.flatten()
yTrainFlat.shape
```

```python
yTestFlat.shape
```

```python
shape = xTestLast.shape[1:]
```

```python
model = models.Sequential()
model.add(layers.InputLayer(input_shape=shape, name="Input"))
model.add(layers.SpatialDropout1D(0.5))
model.add(layers.Conv1D(filters=64, kernel_size=(10), activation='relu', name="Conv"))
model.add(layers.MaxPooling1D(pool_size=16, name="Pool"))
model.add(layers.Conv1D(filters=64, kernel_size=(10), activation='relu', name="Conv2"))
model.add(layers.MaxPooling1D(pool_size=16, name="Pool2"))
model.add(layers.Conv1D(filters=64, kernel_size=(1), activation='relu', name="Conv3"))
model.add(layers.MaxPooling1D(pool_size=2, name="Pool3"))

model.add(layers.Flatten(name="Flat"))
model.add(layers.Dense(431, name="Dense"))
```

```python
model.compile(optimizer='adam',
              loss=tfa.losses.TripletSemiHardLoss(),
              )
```

```python
print(model.summary())
```

```python
history = model.fit(xTrainLast, 
                    yTrainFlat, 
                    epochs=25,
                    validation_split=0.1
                     )
```

```python
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.xlabel('Epoch')
plt.ylabel('loss')
plt.legend(loc='upper right')
```

```python
# save model as JSON 
jsonModel = model.to_json()
with open("model.json", "w") as json_file:
    json_file.write(jsonModel)
# save weights as HDF5
model.save_weights("model.h5")
 
# to load the files later
jsonFile= open('model.json', 'r')
jsonModel = jsonFile.read()
jsonFile.close()
savedModel = models.model_from_json(jsonModel)
# load weights from HDF5
savedModel.load_weights("model.h5")
```

```python
prediction = model.predict(xTestLast)
```
