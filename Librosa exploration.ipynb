```python
import librosa
import librosa.display
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import soundfile as sf
```

```python
# load the audio file 
path = './gravity.ogg'
# the result is a floating point time series representation of the audio
y, sr = librosa.load(path)
```

```python
y # floating point time series
```

```python
sr # sample rate
```

```python
librosa.display.waveplot(y, sr=sr)
```

```python
# apply a FT to the data
Y = librosa.stft(y)
#converting into energy levels(dB)
Ydb = librosa.amplitude_to_db(Y)
```

```python
plt.figure(figsize=(20, 5))
librosa.display.specshow(Ydb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
plt.figure(figsize=(20, 5))
librosa.display.specshow(Y, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
inverseSTFTY = librosa.istft(Y)
InverseYdb = librosa.db_to_amplitude(Ydb)
# after reverting db to amp
inverseTestdBTransform = librosa.istft(InverseYdb)
np.array_equal(y, inverseSTFTY)
```

```python
plt.plot(y)
plt.show()
```

```python
plt.plot(inverseSTFTY)
plt.show()
```

```python
# Write out audio as 24bit PCM WAV
sf.write('original.wav', y, sr, subtype='PCM_24')
sf.write('inverseTest.wav', inverseSTFTY, sr, subtype='PCM_24')
sf.write('inverseTestdBTransform2.wav', inverseTestdBTransform, sr, subtype='PCM_24')
```
